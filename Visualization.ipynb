Importing libraries pandas as numpy


```python
import pandas as pd
import numpy as np
```

Importing dataset using pandas

```python
df = pd.read_csv('Individual financial profiles.csv')
```

**Correcting Datatypes**

```python
df['Account Balance'] = pd.to_numeric(df['Account Balance'].replace('[\$,]', '', regex=True))
df
```

```python
df['Withdrawals'] = pd.to_numeric(df['Withdrawals'].replace('[\$,]', '', regex=True))
df['Deposits'] = pd.to_numeric(df['Deposits'].replace('[\$,]', '', regex=True))
df['Transfers'] = pd.to_numeric(df['Transfers'].replace('[\$,]', '', regex=True))
df['International Transfers'] = pd.to_numeric(df['International Transfers'].replace('[\$,]', '', regex=True))
df['Loan Amount'] = pd.to_numeric(df['Loan Amount'].replace('[\$,]', '', regex=True))
df['Investments'] = pd.to_numeric(df['Investments'].replace('[\$,]', '', regex=True))
```

```python
df['Income Level'] = pd.to_numeric(df['Income Level'].replace('[\$,]', '', regex=True))
```

```python
df['Interest Rate'] = pd.to_numeric(df['Interest Rate'].str.replace('%', ''))
```

```python
df['Risk Tolerance'] = df['Risk Tolerance'].astype('category')
df['Occupation'] = df['Occupation'].astype('category')
```

```python
df['Investment Goals'] = df['Investment Goals'].astype('category')
df['Loan Status'] = df['Loan Status'].astype('category')
```

```python
df['Loan Purpose'] = df['Loan Purpose'].astype('category')
df['Employment Status'] = df['Employment Status'].astype('category')
df['Transaction Description'] = df['Transaction Description'].astype('category')
```

**Creating Derived Columns**

1.Net flow/Cash flow for each individual can be derived from the data.Cash flow represents the movement of money in an out of an individuals account over a period.

```python
df['Cash Flow'] = df['Deposits'] - (df['Withdrawals'] + df['Transfers'] + df['International Transfers'])
df
```

Positive value of cashflow - implies more money coming in than going out or the person is saving money. Negative value of cashflow - implies more money is going out or the person is accumulating debt. Zero - implies a balanced cashflow. money coming in equals money going out.

2.Loan to Income ratio.

```python
df['Loan to Income Ratio'] = df['Loan Amount'] / df['Account Balance']
df
```

Loan to income ratio measures how large a person's loan is in comparison to their income. if the ratio is < 1 - more income than loan-implies low risk if the ratio is >1 - less income than loan - implies high risk if the ratio is =1 - loan equals to income - implies moderate risk.
